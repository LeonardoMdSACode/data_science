```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# getting the data
data = pd.read_csv("train.csv", sep=",")

# plotting the heatmap
corrmat = data.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(15, 15))
sns.set(font_scale=0.5)
heatmap = sns.heatmap(corrmat, annot=True, cmap="RdYlGn")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=60) # rotate the text in order to be lisible
heatmap.set_yticklabels(heatmap.get_yticklabels(), rotation=0) # making sure it is aligned horizontally
plt.show()
```
